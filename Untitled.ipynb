```python
#dependencies
import pandas as pd
```

```python
#loading file
all_disney_movies = 'Resources/disney_movies.csv'
```

```python
#testing file
read_file = pd.read_csv(all_disney_movies)
read_file.head()
```

```python
#Add Release Month and Release Year columns 
read_file['release_date'] = pd.to_datetime(read_file.release_date)
year = read_file['release_date'].dt.strftime('%Y')
year
month = read_file['release_date'].dt.strftime('%m')
read_file['release_year'] = year
read_file['release_month'] = month
read_file.head()
```

```python

clean_file = read_file.dropna(how = 'any')
clean_file
```

```python
#Drop all NaN values
clean_file = clean_file.loc[clean_file['release_year'] >= '1990']
clean_file
```
